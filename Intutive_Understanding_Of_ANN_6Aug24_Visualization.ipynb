```python
import numpy as np
import plotly.graph_objects as go

# Set the random seed for reproducibility
np.random.seed(42)

# Generate a range of values for two variables
W10 = np.linspace(0, 10, 40)
B5 = np.linspace(0, 10, 40)
W10, B5 = np.meshgrid(W10, B5)  # Create a meshgrid for the W10 and B5 values

# Create a complex dataset
MAE = np.sin(W10) * np.log(B5 + 1) + np.cos(B5) * np.exp(W10 / 10) + np.random.normal(0, 0.5, W10.shape)

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(z=MAE, x=W10, y=B5, colorscale='Viridis', opacity=0.9)])

# Example point to highlight on the surface
highlight_x = 5  # Adjust this value as needed
highlight_y = 5  # Adjust this value as needed

# Calculate the function at the specific point
highlight_z = np.sin(highlight_x) * np.log(highlight_y + 1) + np.cos(highlight_y) * np.exp(highlight_x / 10)

# Add the point as a scatter plot marker on the surface
fig.add_trace(go.Scatter3d(
    x=[highlight_x],
    y=[highlight_y],
    z=[highlight_z],
    mode='markers',
    marker=dict(
        size=8,
        color='red',
        symbol='circle'
    )
))

# Update layout for a better view
fig.update_layout(
    title='3D Surface Plot with Highlighted Point',
    scene=dict(
        xaxis_title='W10',
        yaxis_title='B5',
        zaxis_title='MAE'
    ),
    autosize=True,
    width=800,
    height=600,
    margin=dict(l=65, r=50, b=65, t=70)
)

# Show the plot
fig.show()
```
